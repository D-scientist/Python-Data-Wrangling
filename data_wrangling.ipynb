# Data Normalization 
Normalize data is Bell sahpe curve data
- Uniform the data 
- they have same impact 
- Also for computational reasons
- Data should be in one range , therfor we can compare them more better

```python
ks1=boatextra2[['age in days','fare']]
ks1.head()
```

- The above data is really in wide range and we need to normalize and hard to compare

- Normalization change the values to the range of 0-to-1 (now both variable has similar influence on our models)

Method of Normalization

1. Simple feature scaling 
    - x(new)= x(old)/x(max)
2. Min-Max Methods
3. Z-score (standard Score) -3 to +3
4. Log Transformation

```python
# Simple Feature Scaling / Normaliza "Fare" and "age in days"
ks1['fare']= ks1['fare']/ks1['fare'].max() 
ks1['age in days']= ks1['age in days']/ks1['age in days'].max()

ks1.head()
```

```python
# Min - Max Methods 
ks1['fare']= (ks1['fare']-ks1['fare'].min())/(ks1['fare'].max()-ks1['fare'].min())
ks1.head()
```

```python
# Loading Data set
ks2=sns.load_dataset('titanic')
ks2
```

```python
# Z-Score (Standard score)
ks2['fare']= (ks2['fare']-ks2['fare'].mean())/(ks2['fare'].std())
ks2.head()
```

```python
# Log Transformation
ks2['fare'] = np.log(ks2['fare'])
ks2.head()
```

Binning
- Grouping of values into smaller number of values (bins)
- Convert numeric into categories (jawan, Bachay, Boorhay) or 1-16, 17-30 etc
- To have better understanding of groups
    - low vs mid vs high price

```python
bins = np.linspace (min (ks2['age']), max(ks2['age']), 15000) 
agegroups = ["Bachay", "Jawan", "Boorhay"]
ks2['age'] = pd.cut(ks2['age'], bins=bins, labels=agegroups, include_lowest=True)
ks2['age']
```
